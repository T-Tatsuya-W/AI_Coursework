## Used Libraries

```python
import numpy as np
import pandas as pd
```

Now import the datasets from the CSV files into pandas.

```python
dataset = pd.read_csv('./mush.csv')

print("train set : ", dataset.shape)
# dataset
```

```python
columns = dataset.columns


for col in columns:

    #get all unique values in the column
    allValues = dataset[col].unique()
    #print(allValues)
    #iterate through all unique values in the column
    for val in allValues:

        #create the new column name
        nextflagName = col[0]+"_"+val+"_flag"

        #starts creating the new column
        nextColumn = []
        for a in dataset[col]:
            if (a == val):
                nextColumn.append(1)
            else:
                nextColumn.append(0)

        #creates the new column
        dataset[nextflagName] = nextColumn



(dataset)

```
